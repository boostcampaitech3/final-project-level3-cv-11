### Video Version

```python
from tqdm import tqdm
# from paddle import inference
import os, cv2
import numpy as np
from PIL import ImageDraw, ImageFont, Image
import pickle
from detector import detect_faces
from visualization_utils import show_results, draw_face
import warnings
warnings.filterwarnings('ignore')


cap = cv2.VideoCapture('./mp4s/test.mp4') # FIXME
ret = True
frame_width, frame_height = int(cap.get(3)), int(cap.get(4))
size = (frame_width, frame_height)

result = cv2.VideoWriter('./mp4s/result.mp4', 
                        cv2.VideoWriter_fourcc(*'mp4v'),
                        20.0, size)

while ret:
    ret, img = cap.read()
    
    if ret:
        data = Image.fromarray(img)
        boxes, landmarks = detect_faces(data)
        img = draw_face(img, np.array(boxes), landmarks)
        result.write(img)
        if cv2.waitKey(1) & 0xFF == ord('s'):
            break
    else : 
        break
cap.release()
result.release()    

cv2.destroyAllWindows()
print('finished')
```
